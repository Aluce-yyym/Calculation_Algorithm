```python
def f(x):
    return -x**3 + 3*x + 2

def dfdx(x):
    return -3*(x**2) + 3

def forward_diff(a, h):  # 前進差分
    return (f(a + h) - f(a)) / h

def backward_diff(a, h):  # 後退差分
    return (f(a) - f(a - h)) / h

def central_diff(a, h):  # 中心差分
    return (f(a + h) - f(a - h)) / (2 * h)

def relative_error(num, a):
    return abs(num - dfdx(a)) / abs(dfdx(a))


h = 0.1
a = 0.5
print(f"h={h}, a={a}のとき")

forward_diff_val = forward_diff(a, h)
print(f"前進差分による数値微分: {round(forward_diff_val, 2)}")
forward_diff_rel_error = relative_error(forward_diff_val, a)
print(f"前進差分の相対誤差: {round(forward_diff_rel_error, 3)}")

backward_diff_val = backward_diff(a, h)
print(f"後退差分による数値微分: {round(backward_diff_val, 2)}")
backward_diff_rel_error = relative_error(backward_diff_val, a)
print(f"後退差分の相対誤差: {round(backward_diff_rel_error, 3)}")

central_diff_val = central_diff(a, h)
print(f"中心差分による数値微分: {round(central_diff_val, 2)}")
central_diff_rel_error = relative_error(central_diff_val, a)
print(f"中心差分の相対誤差: {round(central_diff_rel_error, 3)}")


```

```python
def f(x):
    return x**4 + x**3 + x**2 + x + 1

def dfdx(x):
    return 4*x**3 + 3*x**2 + 2*x + 1

def forward_diff(a, h):  # 前進差分
    return (f(a + h) - f(a)) / h

def backward_diff(a, h):  # 後退差分
    return (f(a) - f(a - h)) / h

def central_diff(a, h):  # 中心差分
    return (f(a + h) - f(a - h)) / (2 * h)

def relative_error(num, a):
    return abs(num - dfdx(a)) / abs(dfdx(a))


h = 0.1
a = 0.5
def result(h, a):
    print(f"h={h}, a={a}のとき")

    forward_diff_val = forward_diff(a, h)
    print(f"前進差分による数値微分: {round(forward_diff_val, 2)}")
    forward_diff_rel_error = relative_error(forward_diff_val, a)
    print(f"前進差分の相対誤差: {round(forward_diff_rel_error, 3)}")

    backward_diff_val = backward_diff(a, h)
    print(f"後退差分による数値微分: {round(backward_diff_val, 2)}")
    backward_diff_rel_error = relative_error(backward_diff_val, a)
    print(f"後退差分の相対誤差: {round(backward_diff_rel_error, 3)}")

    central_diff_val = central_diff(a, h)
    print(f"中心差分による数値微分: {round(central_diff_val, 2)}")
    central_diff_rel_error = relative_error(central_diff_val, a)
    print(f"中心差分の相対誤差: {round(central_diff_rel_error, 3)}")


for i in range(10):
    result(h = 0.1, a = i*0.1)
    print()
```
